### Part a) Show that: 
$$ 
\frac{\partial \mathbf{a^Tx}}{\partial \mathbf x} = \mathbf a^T 
$$

For an element \( k \):
$$
\frac{\partial \mathbf{a^Tx}}{\partial x_k} = \frac{\partial }{\partial x_k} \sum_{i=1}^{N} a_i \cdot x_i = \sum_{i=1}^{N} a_i \cdot \frac{\partial x_i}{\partial x_k} = a_k 
$$

Thus,
$$ 
\frac{\partial \mathbf{a^Tx}}{\partial \mathbf x} = [a_1, a_2, \ldots, a_N] = \mathbf a^T \blacksquare 
$$

### Part b) Show that: 
$$
\frac{\partial (\mathbf{a^TAa})}{\partial \mathbf a} = \mathbf a^T \left( A + A^T \right)
$$

For an element \( k \):
$$
\frac{\partial (\mathbf{a^TAa})}{\partial a_k} = \frac{\partial }{\partial a_k} \sum_{i=1}^{N} \sum_{j=1}^{N} a_i A_{ij} a_j 
$$

$$
= \sum_{i=1}^{N} \sum_{j=1}^{N} \left( \frac{\partial a_i}{\partial a_k} A_{ij} a_j + a_i A_{ij} \frac{\partial a_j}{\partial a_k} \right) 
= \sum_{j=1}^{N} A_{kj} a_j + \sum_{i=1}^{N} a_i A_{ik} 
$$

$$
= \sum_{i=1}^{N} a_i \left( A_{ik} + A_{ki} \right) 
$$

Thus,
$$
\frac{\partial (\mathbf{a^TAa})}{\partial \mathbf x} = \mathbf a^T \left( A + A^T \right) \blacksquare
$$

### Part c) Show that:
$$
\frac{\partial (\mathbf{x-As})^T(\mathbf{x-As})}{\partial \mathbf s} = -2 \left(\mathbf{x-As} \right)^T \mathbf A
$$

For an element \( k \):
$$
\frac{\partial }{\partial s_k} \sum_{i=1}^{N} \left( x_i - \sum_{j=1}^{N} A_{ij} s_j \right)^2 
$$

$$
= \sum_{i=1}^{N} \left( -2x_i \sum_{j=1}^{N} A_{ij} \frac{\partial s_j}{\partial s_k} + 2 \left( \sum_{j=1}^{N} A_{ij} \frac{\partial s_j}{\partial s_k} \right) \left( \sum_{j=1}^{N} A_{ij} s_j \right) \right)
$$

$$
= \sum_{i=1}^{N} \left( -2x_i A_{ik} + 2 A_{ik} \left( \sum_{j=1}^{N} A_{ij} s_j \right) \right)
$$

$$
= -2 \sum_{i=1}^{N} \left( x_i - \sum_{j=1}^{N} A_{ij} s_j \right) A_{ik}
$$

Thus,
$$
\frac{\partial (\mathbf{x-As})^T(\mathbf{x-As})}{\partial \mathbf s} = -2 \left( \mathbf{x-As} \right)^T \mathbf A \blacksquare
$$

### Part d) Show that:
$$
-2\frac{\partial \left( (\mathbf{x-As})^T \mathbf A \right)}{\partial \mathbf s} \propto A^T A
$$

First of all,
$$
-2\frac{\partial \left( (\mathbf{x-As})^T \mathbf A \right)}{\partial \mathbf s} = -2\frac{\partial \left( \mathbf{x}^T \mathbf A - \mathbf{(As)}^T \mathbf A \right)}{\partial \mathbf s} = 2 \frac{\partial \left( \mathbf{(As)}^T \mathbf A \right)}{\partial \mathbf s}
$$

For an element \( l, k \):
$$
2\frac{\partial \left( \mathbf{(As)}^T \mathbf A \right)_l}{\partial s_k} = 2\sum_{i=1}^{N} A_{il} \sum_{j=1}^{N} A_{ij} \frac{\partial s_j}{\partial s_k} = 2\sum_{i=1}^{N} A_{il} A_{ik} = 2\sum_{i=1}^{N} A_{il} A_{ki}^T
$$

Thus,
$$
-2\frac{\partial \left( (\mathbf{x-As})^T \mathbf A \right)}{\partial \mathbf s} = 2A^T A \blacksquare
$$

<h2 style="font-family:Times New Roman; font-size:200%">Generating polynomial data</h2>


```julia
n=100
f(x) = 2x^2-3*x-2
#x_data = randn(n) #Normally distributed x-data, this could of course be linear as well, it doesn't really matter
x_data = LinRange(-3,3, 100) |> collect # Linear
y_data = f.(x_data).+ 0.1.*randn(n)
```

```julia
scatter(x_data,y_data, title="Polynomial Data", label="Data")
```

<h2 style="font-family:Times New Roman; font-size:200%">Splitting the data into test/train</h2>


```julia
function SplitData(xData::Array, yData::Array, train_size::Float64)
    #Assumes Data : n_features, n_observations
    
    if ndims(xData)==1
        N = length(xData)
        NumTrain = Int(round(N*train_size))
            
        train_indices = sample(1:N, NumTrain; replace=false)
        all_indices = collect(1:N)
        test_indices = [i for i in all_indices if i ∉ train_indices]
        
        xTrain = xData[train_indices]
        yTrain = yData[train_indices]
        xTest = xData[test_indices]
        yTest = yData[test_indices]
        
    elseif ndims(xData)==2 #If there are for multiple features for each datapoint i.e position, velocity
        N = length(xData[1,:])
        NumTrain = Int(round(N*train_size))
            
        train_indices = sample(1:N, NumTrain; replace=false)
        all_indices = collect(1:N)
        test_indices = [i for i in all_indices if i ∉ train_indices]
        
        xTrain = xData[:,train_indices]
        yTrain = yData[:,train_indices]
        xTest = xData[:,test_indices]
        yTest = yData[:,test_indices]
    else #This is undefined
        @warn "Array dimensions do not match the expected dimensions." 
        println("Expected: 1 or 2, but got: ", ndims(xData))
        return
    end
    return xTrain, xTest, yTrain, yTest
end
```

```julia
xtrain, xtest, ytrain, ytest = SplitData(x_data, y_data, 0.8)
```

```julia
scatter(xtrain, ytrain, label= "Train", title="Tran/Test split for polynomial")
scatter!(xtest, ytest, label="Test")
```

```julia
function Regression(xData, yData, numDegrees::Int64)
    DesignMatrix = zeros(length(xData), numDegrees+1)
    for i in 1:numDegrees+1
        DesignMatrix[:,i].=xData.^(i-1)
    end
    Hessian = Transpose(DesignMatrix)*DesignMatrix
    beta = inv(Hessian)*Transpose(DesignMatrix)*yData
    return beta
end

function MSE(y,x, beta)
    N = length(x)
    DesignMatrix = zeros(length(x), length(beta)) 
    for i in 1:length(beta)
        DesignMatrix[:,i].=x.^(i-1)
    end
    MSE = 1/N*sum((y .- DesignMatrix*beta).^2)
    return MSE
end

function R2(y,x, beta)
    N = length(x)
    DesignMatrix = zeros(length(x), length(beta)) 
    for i in 1:length(beta)
        DesignMatrix[:,i].=x.^(i-1)
    end
    y_avg = 1/N*sum(y)
    R2 = 1- sum((y.-DesignMatrix*beta).^2)/sum((y .- y_avg).^2)
    return R2
end
```

```julia
betaVals = Regression(xtrain,ytrain, 2)
beta0, beta1, beta2 = betaVals
model(x) = beta0 + beta1*x + beta2*x^2
```

```julia
y_hat=model.(xtest)
println("MSE: ", MSE(ytest,xtest,betaVals))
println("R2 value: ", R2(ytest,xtest,betaVals))
```

```julia
scatter(x_data, y_data, label = "ground truth", title="Result from the model")
scatter!(xtest,y_hat, label = "predicted data")
```

<h2 style="font-family:Times New Roman; font-size:200%">Non polynomial data</h2>


```julia
n=300
f_NotPoly(x) = 2*ℯ^x-5*cos(3*x-2)+3*x+2
#x_data = randn(n) #Normally distributed x-data, this could of course be linear as well, it doesn't really matter
x_data = LinRange(-3,3, n) |> collect # Linear
y_data = f_NotPoly.(x_data).+ 0.1.*randn(n)
scatter(x_data,y_data, label="Data", title="Random non polynomial funciton")
```

```julia
xtrain, xtest, ytrain, ytest = SplitData(x_data, y_data, 0.8)
```

```julia
function RandomPolyDegModel(xData, betas)
    #This function will calculate ỹ = ∑ β_i x^i for any number of degrees 
    N = length(xData)
    DesignMatrix = zeros(length(xData), length(betas)) 
    for i in 1:length(betas)
        DesignMatrix[:,i].=xData.^(i-1)
    end
    est = DesignMatrix*betas
    return est
end
```

```julia
betaVals = Regression(xtrain,ytrain, 1)
y_hat=RandomPolyDegModel(xtest, betaVals)

mse = round(MSE(ytest,xtest,betaVals),digits=2)
r2 = round(R2(ytest,xtest,betaVals),digits=2)

p = plot(x_data, y_data, label = "Ground truth",alpha=0.7,layout=8, size = (900,900),linewidth=3)
scatter!(subplot=1, xtest, y_hat, label = "Degree=2, MSE: $(mse), R2: $(r2)")

degrees = range(2,8) |> collect
for deg in degrees
    betaVals = Regression(xtrain,ytrain, deg)

    y_hat=RandomPolyDegModel(xtest, betaVals)
    
    mse = round(MSE(ytest,xtest,betaVals),digits=2)
    r2 = round(R2(ytest,xtest,betaVals),digits=2)
    
    plot!(subplot=deg, x_data, y_data, label = "Ground truth", alpha=0.7, size = (900,900),linewidth=3)
    scatter!(subplot=deg, xtest,y_hat, label = "Degree=$(deg), MSE: $(mse), R2: $(r2)")
end
display(p)
```
